## Basic Usage

The following demonstrates a simple way to use trace to assemble a scene and trace rays through it.

```python
# Make a scene and include a lense
scene = tr.Scene()

# Circular lense at origin, pointing upwards
lense = tr.SphereLens(1, 1, 1, pos=tr.pos(0, 0, 0), axis=tr.vec(0, 0, 1), n=2.0)
screen = tr.Screen(pos=tr.pos(-1,-1,-2), width=tr.vec(2, 0, 0), height=tr.vec(0, 2, 0))
lamp = tr.Source(tr.pos(0, 0, 2), k=tr.vec(0, 0, -1), R=0.2, N=100)
```

```python
# Add the elements to the simulation and run
scene.add(lense, screen, lamp)
scene.trace()
```

```python
tr.graphics.render_3d(scene, extend=3.0)
tr.graphics.render_2d(screen)
plt.show()
```
