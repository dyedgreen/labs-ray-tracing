# Trace Demo

```python
import numpy as np
import matplotlib.pyplot as plt

import trace as tr
```

```python
# Make a scene and include a lense
scene = tr.Scene()

# Circular lense at origin, pointing upwards
lense = tr.Sphere(1, 1, 1, pos=tr.pos(0, 0, 0), axis=tr.vec(0, 0, 1))
lamp = tr.Source(tr.pos(0, 0, 2), k=tr.vec(0, 0, -1), R=0.2, N=100)
```

```python
# Add the elements to the simulation and run
scene.add(lense, lamp)
scene.trace()
```

```python
tr.graphics.render_3d(scene, extend=2.0)
plt.show()
```
